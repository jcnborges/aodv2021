```python
#!/usr/bin/env python
"""
Plot multi-graphs in 3D.
"""
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import LayeredNetworkGraph as lng

from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Line3DCollection

# define graphs
n = 5
g = nx.erdos_renyi_graph(4*n, p=0.1)
h = nx.erdos_renyi_graph(3*n, p=0.2)
i = nx.erdos_renyi_graph(2*n, p=0.4)

node_labels = {nn : str(nn) for nn in range(4*n)}

# initialise figure and plot
fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.add_subplot(111, projection='3d')
lng.LayeredNetworkGraph([g, h, i], ax=ax, node_labels=node_labels, layout=nx.spring_layout)
ax.set_axis_off()
plt.show()
```
