```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from numpy import mean
from numpy import std
from scipy.stats import norm
%matplotlib inline
```

```python
aodv = pd.read_csv('aodv.csv', sep = ';')
```

```python
aodv.head()
```

```python
nodes = aodv[aodv.destino == '127.0.0.0']['origem']
```

```python
nodes.head()
```

```python
edges = aodv[aodv.timestamp != 0][['acao', 'timestamp', 'origem', 'destino', 'gateway']]
```

```python
edges.head()
```

```python
G = nx.DiGraph()
for n in nodes:
    G.add_node(n)
start = edges.iloc[0]['timestamp']
offset = 1.8e9
for idx, e in edges[edges.timestamp <= start + (offset / 1000)].iterrows():
    if '+' in e['acao']:
        if not G.has_edge(e['origem'], e['gateway']):
            G.add_edge(e['origem'], e['gateway'])
nx.draw(G, pos=nx.kamada_kawai_layout((G)))
dg = []
for n, d in nx.degree(G):
    dg.append(d)
print('Offset: {0} ms'.format(offset))
print('Fracamente conectado: {0}'.format(nx.is_weakly_connected(G)))
print('Fortemente conectado: {0}'.format(nx.is_strongly_connected(G)))
print('Densidade: {0}'.format(nx.density(G)))
try:
    print('Menor caminho médio: {0}'.format(nx.average_shortest_path_length(G)))
except:
    print('Menor caminho médio: -')
try:
    print('Diâmetro: {0}'.format(nx.diameter(G)))
except:
    print('Diâmetro: -')
```

```python
media = mean(dg)
sigma = std(dg)
dist = norm(media, sigma)
valores = range(0, max(dg) + 5)
prob = [dist.pdf(x) for x in valores]
plt.hist(x=dg, bins=10, align='left', density=True)
plt.plot(valores, prob)
plt.show()
```
